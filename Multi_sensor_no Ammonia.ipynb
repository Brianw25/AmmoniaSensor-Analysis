```python
import numpy as np
import scipy.constants as const
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from datetime import date
from scipy.ndimage import gaussian_filter1d
%matplotlib inline
```

```python
dat = pd.read_excel('NoAM.xlsx')
dat.head()
```

```python
#dat = dat.loc[0:2500]
dat['Sensor 1'] = gaussian_filter1d(dat['Sensor 1'], 5)
dat['Sensor 2'] = gaussian_filter1d(dat['Sensor 2'], 5)
dat['Sensor 3'] = gaussian_filter1d(dat['Sensor 3'], 5)
dat['Sensor 4'] = gaussian_filter1d(dat['Sensor 4'], 5)
dat.plot(subplots=True, layout=(2,2))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20,20),constrained_layout=True)
ax1.grid()
ax1.plot(dat.index ,dat['Sensor 1']/1000,'-',label = 'Sensor Resistance', linewidth = 5)#Sensor
ax1.set_title('Sensor 1 Response', fontsize = 32, fontweight ='bold')
ax1.set_ylabel('Resistance (k\u03A9)',fontsize = 22)
ax1.set_xlabel('Time (sec)',fontsize = 22)

ax2.grid()
ax2.plot(dat.index ,dat['Sensor 2']/1000,'-',label = 'Sensor Resistance', linewidth = 5)#Sensor
ax2.set_title('Sensor 2 Response', fontsize = 32, fontweight ='bold')
ax2.set_ylabel('Resistance (k\u03A9)',fontsize = 22)
ax2.set_xlabel('Time (sec)',fontsize = 22)

ax3.grid()
ax3.plot(dat.index ,dat['Sensor 3']/1000,'-',label = 'Sensor Resistance', linewidth = 5)#Sensor
ax3.set_title('Sensor 3 Response', fontsize = 32, fontweight ='bold')
ax3.set_ylabel('Resistance (k\u03A9)',fontsize = 22)
ax3.set_xlabel('Time (sec)',fontsize = 22)

ax4.grid()
ax4.plot(dat.index ,dat['Sensor 4']/1000,'-',label = 'Sensor Resistance', linewidth = 5)#Sensor
ax4.set_title('Sensor 4 Response', fontsize = 32, fontweight ='bold')
ax4.set_ylabel('Resistance (k\u03A9)',fontsize = 22)
ax4.set_xlabel('Time (sec)',fontsize = 22)
```
